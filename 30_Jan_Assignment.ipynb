```python
#Q2 using for loop
sum = 0
prod = 1
for i in range (1,11):
  sum += i
  prod *= i
print("sum is " , sum)
print("product is ", prod)
```

```python
#Q2 using while loop
sum = 0
prod = 1
i = 1
while i <= 10:
  sum += i
  prod *= i
  i += 1
print("sum is " , sum)
print("product is ", prod)
```

```python
#Q3
units = int(input("Enter units of Electricity consumed"))
cost = 0
while units > 0:
  if units >= 100:
    rem = units - 100
    units = units - 100
    if (rem < 0):
      cost = rem*4.5 + cost
    else:
      cost = cost + 4.5*100
  if units > 0:
    rem = units - 100
    units = units - 100
    if (rem < 0):
      cost = rem*6 + cost
    else:
      cost = cost + 6*100
  if units > 0:
    rem = units - 100
    units = units - 100
    if (rem < 0):
      cost = rem*10 + cost
    else:
      cost = cost + 10*100

  if units > 0:
    cost = units*20 + cost
    units = 0


print("cost is" , cost)
```
